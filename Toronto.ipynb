```python
import pandas as pds
pds_df = pds.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
pds_df[0].head()
```

```python
assigned_df=pds_df[0][pds_df[0].Borough!='Not assigned']
assigned_df.size
```

```python
assigned_df.query("Postcode=='M5A'")
```

```python
grouped_df = assigned_df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(lambda tags: ','.join(tags))
grouped_df.head()
```

```python
coll_df=grouped_df.reset_index()
coll_df.head()
```

```python
coll_df.Neighbourhood.replace('Not assigned',coll_df['Borough'],inplace=True)
coll_df
```

```python
geo_df = pds.read_csv('Geospacial_Coordinates.csv')
geo_df.head()
```

```python
geo_df.columns = geo_df.columns.str.replace('Postal Code','PostalCode')
geo_df.columns
```

```python
coll_df.columns = coll_df.columns.str.replace('Postcode','PostalCode')
coll_df.columns
```

```python
merged_df = pds.merge(coll_df,geo_df,on='PostalCode')
merged_df.head()
```

Create a map of Toronto with neighborhoods superimposed on top.


```python
#!conda install -c conda-forge folium=0.5.0 --yes
```

```python
import folium
```

```python
# create map of New York using latitude and longitude values
latitude, longitude=43.687829, -79.407643
map_Toronto = folium.Map(location=[latitude, longitude], zoom_start=11)
# add markers to map
for idx, post, borough, neighborhood, lat, lng in merged_df.itertuples():
    label = '{}, {}'.format(post, neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_Toronto)
map_Toronto
```

Feel free to zoom into the above map, and click on each circle mark to reveal the name of the neighborhood and its respective borough.
